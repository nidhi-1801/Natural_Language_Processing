**Lemmatizer**

Lemmatization technique is like stemming. which reduces the word into it's root form rather than stem word like stemming. After lemmatization, we will be getting a valid word that means something.


**WordNet Lemmatizer**

NLTK provides Wordnet Lemmatizer,It uses WordNet lexical database to find the lemma based on POS tags

USECASE:
chatbots, text summarization, Q&A


```python
words = ["studies", "playing", "better", "mice"]
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("going")  #the root form is go but it gave going. By default, NLTK assumes every word is a noun (POS = “n”) unless you tell it otherwise.
```

```python
'''
pos
Noun - n
Verb - v
adjective - a
adverb - r
'''
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("going", pos = "v")
```

```python
lemmatizer = WordNetLemmatizer()
for w in words:
 print(lemmatizer.lemmatize(w, pos='v'))
```

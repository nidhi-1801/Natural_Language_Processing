```python
Paragraph = ''' Dream, dream, dream.
Dreams transform into thoughts.
Thoughts result in action.

You have to dream before your dreams can come true.

When I was a young boy, I had a dream to fly.
That dream led me to study aeronautical engineering.
It led me to work on India's space programs.
It finally led me to become the President of India.

Each one of you has hidden potential.
Success comes to those who work hard and stay confident.

You must have a clear goal in life.
Obstacles will appear, but they test your determination.

If you fail, never give up.
FAIL means "First Attempt In Learning".
END is not the end; it means "Effort Never Dies".
If you get NO as an answer, remember it means "Next Opportunity".

Dream big, think big, and work hard.
That is the way to achieve success. '''
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
#Step1, tokenize the text
import nltk
from nltk.tokenize import word_tokenize
tokens = word_tokenize(Paragraph)
```

```python
#Step2, Apply post tag
pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

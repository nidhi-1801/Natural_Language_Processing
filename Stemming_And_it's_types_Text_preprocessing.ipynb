**STEMMING**

stemming is the process of reducing words to their root or base form by removing suffixes or prefixes.

Disadvantage: When it tries to reduce to root form sometime, it give the words which are not meaningful


```python
#Classification problem
#Comment of product is a posiitive or negative review
#Reviwes ----> eating,eat,eaten,going,goes,go..............
```

Type1: **PorterStemmer**

Most commonly used; applies a series of rules to remove suffixes and prefixxes

```python
words = ["playing", "played", "player", "studies", "studying","historical"]
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
for w in words:
  print(porter.stem(w))    #here it gave histor which doesnt mean anything which is a disadvantage
```

**RegexpStemmer**

Uses regular expressions to manually define suffix patterns to remove.

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$',min=4)
for w in words:
  print(reg_stemmer.stem(w))
```

**SnowballStemmer**

This is refined version of PorterStemmer. Perfoms better

```python
from nltk.stem import SnowballStemmer
Snowball = SnowballStemmer('english')   # You need to specific english as snowballstememr supports many languages
for w in words:
  print(Snowball.stem(w))
```

```python
#Demosntrates SnowballStemmer performs better than PorterStemmer
from nltk.stem import SnowballStemmer
a = SnowballStemmer('english')
print(a.stem("fairly"))

from nltk.stem import PorterStemmer
a = PorterStemmer()
print(a.stem("fairly"))
```

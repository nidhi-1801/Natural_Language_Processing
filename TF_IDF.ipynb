```python
corpus = [
    "I love machine learning",
    "machine learning is fun",
    "I love deep learning"
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
x = tf.fit_transform(corpus).toarray()
print(tf.get_feature_names_out())
print(x)
```
